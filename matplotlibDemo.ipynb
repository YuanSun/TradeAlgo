# Matplotlib Basics
- It is the most powerful and customizable visualizatino library for Python.
- It was initially based off Matlab's plotting capabilies.
- To get an idea of what is possible, go to http://matplotlib.org/gallery.html
- Understanding its core principles will be very helpful later on with pandas.

```python
import matplotlib.pyplot as plt
%matplotlib inline 
# This is to allow you to see the plot you create inside notebook
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# OBJECT ORIENTED METHOD
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.9])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes2.plot(y, x)

axes2.set_title('SMALLER PLOT')
axes1.set_title('LARGER PLOT')
```

# Basic Part 2

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
# in this case, axes is an array that you can iterate
plt.tight_layout() # solve the overlapping issue
# for current_ax in axes:
#     current_ax.plot(x, y)

axes[0].plot(x, y)
axes[1].plot(y, x)

axes[0].set_title('First plot')
axes[1].set_title('Second plot')
```

# Figure Size and DPI

```python
fig = plt.figure(figsize = (8, 2), dpi=150)
ax = fig.add_axes([0,0,1,1])
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize=(8, 2))
axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

```python
fig.savefig('my_pic.png', dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='X Squared')
ax.set_title('Title')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

ax.plot(x, x**3, label='X Cubed') # plot multiple figs in one canvas

ax.legend(loc='best')
```

# Matplotlib - Part three

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='#FF8C00', linewidth=1, alpha=1, linestyle='-', marker='o', markersize=20, 
        markerfacecolor='green', markeredgewidth=3, markeredgecolor='red') 
# RGB Hex Code, alpha is transparent degree

```

```python
fig = plt.figure()
ax2=fig.add_axes([0,0,1,1])
ax2.plot(y, y**2)

ax2.set_xlim([0,1])
ax2.set_ylim([0,2])
```
